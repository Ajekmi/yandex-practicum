## Предобработка данных

Выведем столбцы с пропусками:

```python
data.isna().sum()
```

Дальнейший план по замене пропусков: 
* ceiling_height - заменим пропуски медианной по выбранным группам
* floors_total - удалим пустые значения, так как доля пропусков очень маленькая
* living_area - заполним средней долей living_area в столбце total_area
* is_apartment -  заменим незаполненные значения на False
* kitchen_area - заполним средней долей kitchen_area в столбце total_area
* balcony - заменим пропуски на нулевые значения, т.к. скорее всего они означают отсутствие балкона
* locality_name - удалим пустые значения, так как доля пропусков очень маленькая
* airports_nearest - не можем заполнить 
* cityCenters_nearest - заменим то, что возможно медианой высчитанной по населенным пунктам 
* parks_around3000 - заполним пропущенные данные 0, так как скорее всего пропуски означают, что парка в радуисе 3км нет
* parks_nearest - не будем заполнять такое большое количество пустых данных
* ponds_around3000 - заполним пропущенные данные 0, так как скорее всего пропуски означают, что пруда в радуисе 3км нет
* ponds_nearest -  не будем заполнять такое большое количество пустых данных
* days_exposition - заменим медианой по столбцу.

Изменим тип данных столбца `last_price` на целочисленный:

```python
data['last_price'] = data['last_price'].astype('int')
```

Заменим пропуски в `balcony` на значение 0, потому что скорее всего эти пропуски означают отсутствие балкона и изменим формат данных, чтобы число балконов было целочисленным:

```python
data['balcony'] = data['balcony'].fillna(value = 0).astype('int')
```

Изменяем тип данных в столбце `first_day_exposition`на datetime, так как сейчас это тип object, чтобы в дальнейшем работать с датами и добавить в таблицу день недели, месяц и год публикации объявления.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')
```

В столбце `is_apartment` очень много пропущенных значений, скорее всего этот столбец не заполнялся, когда объект не являлся апартаментами, поэтому заменим незаполненные занчения на False.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Так как у нас не известны занчения столбцов `living_area` и `kitchen_area`, но известна общая площадь `total_area`, попробуем заменить неизвестные значения столбцов средней долей жилой площади и кухни в общей площади:

```python
part_living_area = data['living_area'].median() / data['total_area'].median()
# data['living_area'].fillna(part_living_area*data['total_area'], inplace=True) <-- то, что изменяла
```

```python
data['living_area'] = data['living_area'].fillna(part_living_area*data['total_area'])
```

```python
part_kitchen_area = data['kitchen_area'].median() / data['total_area'].median()
# data['kitchen_area'].fillna(part_kitchen_area*data['total_area'], inplace=True) <-- то, что изменяла
```

```python
data['kitchen_area'] = data['kitchen_area'].fillna(part_kitchen_area*data['total_area'])
```

Столбец с расстоянием объекта недвижимости до центра `cityCenters_nearest` заменим медианой по группам и переведем в целочисленный формат: 

```python
data['cityCenters_nearest'] = data['cityCenters_nearest'].fillna(data.groupby('locality_name')['cityCenters_nearest'].transform('median'))
```

Заполним пустые значения столбцов `parks_around3000` и `ponds_around3000` на 0 и переведем формат в целочисленный:

```python
data['parks_around3000'] = data['parks_around3000'].fillna(value = 0).astype('int')
```

```python
data['ponds_around3000'] = data['ponds_around3000'].fillna(value = 0).astype('int')
```

Заменим медианой по столбцу незаполненные значения столбца `days_exposition`:

```python
data['days_exposition'] = data['days_exposition'].fillna(value = data['days_exposition'].median()).astype('int')
```

Удалим пустые значения столбцов `floors_total` и `locality_name`, переведём значения столбца `floors_total` в формат int:

```python
data.dropna(subset=['floors_total'], inplace=True)
data.dropna(subset=['locality_name'], inplace=True)
data['floors_total']=data['floors_total'].astype('int')
```

Заменим пропущенные значения столбца `ceiling_height` с помощью метода transform, благодаря которому в каждой строке содержится медиана посчитанная по заданной группе. После этого вместо пропусков в оригинальном столбце мы подставляем значения из медианы в зависимости от группы.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby(['floors_total', 'locality_name'])['ceiling_height'].transform('median'))
```

```python
part_missing_ceiling_height = data['ceiling_height'].isna().mean()
print(' Доля пропущенных значений высоты потолков {:.1%}'.format(part_missing_ceiling_height))
```

В данных ещё осталось 1.7% пропущенных значений столбца `ceiling_height`, но они не влияют на дальнейшее выполнение проекта.

Переводим столбец `floors_total` в целочисленный формат, так как в дальнейшем будем производить с ним операции:

```python
data['floors_total'] = data['floors_total'].astype('int')
```

Переводим столбец `days_exposition` в целочисленный формат, так как в дальнейшем будем производить с ним операции:

```python
data['days_exposition'] = data['days_exposition'].astype('int')
```

## Расчёты и добавление результатов в таблицу

Расчет и добавление в таблицу цены квадратного метра:

```python
data['price_sq_m'] = (data['last_price'] / data['total_area']).astype('int')
```

Добавление в таблицу день недели, месяц и год публикации объявления:

```python
data['weekday'] = pd.DatetimeIndex(data['first_day_exposition']).weekday
```

```python
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month
```

```python
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

Расчет и добавление в таблицу этажа квартиры по категориям: первый, последний, другой:

```python
def floor_category(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
data['floor_category'] = data.apply(floor_category, axis=1)
```

Расчет и добавление в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей:

```python
data['ratio_living'] = (data['living_area'] / data['total_area'] * 100).astype('int')
```

```python
data['ratio_kitchen'] =  (data['kitchen_area'] / data['total_area'] * 100).astype('int')
```

###  Удаление редких и выбивающихся значений.

Проверим основные параметры талбицы:

```python
data.describe()
```

Уберем аномально высокие заначения стоимости, общей площади объекта недвижимости, жилой площади, площади кухни, количества комнат и высоты потолков функцией для подсчёта границ с учетом 1.5 квартильного размаха.

```python
def subset_by_iqr(df, column):  # <-- то, что я добавила
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    # Возвращаем датафрейм с обрезанными выбросами с обеих сторон в заданном столбце
    res = df.loc[(df[column] > (q1 - 1.5*iqr)) & (df[column] < (q3 + 1.5*iqr)), column]
    part_missing = data[column].isna().mean()
    print(f'Доля удаленных данных столбца {column} =' + ' {:.1%}'.format(part_missing))
          
    return res
```

```python
for column in ['rooms', 'ceiling_height', 'total_area', 'last_price', 'living_area', 'kitchen_area']: # <-- то, что я добавила
    data[column] = subset_by_iqr(data, column)
```
